```python
def phi1_3(x_arr, x_phi_1 = x_phi_1, phi_0 = phi_0): #cauttion, if x = a node, won't work :(
  # needs to be zero dimensional
  x_arr = np.reshape(x_arr, (len(x_arr)))

  x_phi_np = (x_phi_1.numpy()).flatten()
  results = []


  # Find the indices where x_1 would fit in the sorted array x
  for x in x_arr:
    idx_left = np.searchsorted(x_phi_np, x, side='left')
    idx_right = np.searchsorted(x_phi_np, x, side='right')

    if x <= x_phi_np[0]: #if x is out of range, use first two indecies
      idx_left = 0
      idx_right = 1
    elif x >= x_phi_np[-1]: #if x is out of range, use last two indecies
      idx_left = len(x_phi_np) - 2
      idx_right = len(x_phi_np) - 1
    else:
      idx_left = max(idx_left -1,0)                  # idx_left-1, i had this before below, but this allows interp correct dimensions
      idx_right = min(idx_right, len(x_phi_np) - 1)


    # Get the interval [x[idx_left-1], x[idx_right]]
    interval = [x_phi_np[idx_left], x_phi_np[idx_right]]

    # h for x in that that interval
    int_diff = interval[1] - interval[0] # thiis is positive value

    # using relu. figured out the linear comb
    interp = (np.transpose(phi_0)[idx_left] * (tf.nn.relu(x - x_phi_1[idx_left:idx_right])/(-int_diff) + 1) +
              np.transpose(phi_0)[idx_right] * (tf.nn.relu(x - x_phi_1[idx_left:idx_right])/int_diff))

    results.append(tf.reduce_sum(interp).numpy())
    #results.append(tf.reduce_sum(interp))
  # get a numpy array
  results = tf.constant(results)





  return results
```

```python
# sorted takes things that need to be in order!!!
def phi1_4(x_arr, x_phi_1 = x_phi_1, phi_0 = phi_0):
    # Convert inputs to TensorFlow tensors if they aren't already
    x_arr = tf.convert_to_tensor(x_arr, dtype=tf.float64)
    x_phi_1 = tf.convert_to_tensor(x_phi_1, dtype=tf.float64)
    phi_0 = tf.convert_to_tensor(phi_0, dtype=tf.float64)
    #print(phi_0.shape)
    # Flatten the x_phi_1 tensor
    x_phi_np = tf.reshape(x_phi_1, [-1])
    #print(x_phi_np[0:2])
    #print(x_phi_np[-2:])
    #print(x_phi_np.shape)
    results = []

    # Find the indices where x_1 would fit in the sorted array x
    for x in tf.unstack(x_arr):
        idx_left = tf.cast(tf.searchsorted(tf.constant(x_phi_np), tf.constant(x, dtype=tf.float64), side='left'), tf.int64)
        #print(f'left index:, {idx_left}')
        idx_right = tf.cast(tf.searchsorted(tf.constant(x_phi_np), tf.constant(x, dtype=tf.float64), side='right'), tf.int64)
        #print(f'right index:, {idx_right}')

        if tf.constant(x, dtype=tf.float64) <= x_phi_np[0]:  # if x is out of range, use first two indices
            idx_left = 0
            idx_right = 1
        elif tf.constant(x, dtype=tf.float64) >= x_phi_np[-1]:  # if x is out of range, use last two indices
            idx_left = len(x_phi_np) - 2
            idx_right = len(x_phi_np) - 1
        else:
            idx_left = tf.maximum(idx_left - 1, 0)
            #print(f'left index:, {idx_left}')
            idx_right = tf.minimum(idx_right, len(x_phi_np) - 1)

        # Get the interval [x[idx_left], x[idx_right]]
        #print('this is x_phi_1[left_indx]')
        #print(tf.gather(x_phi_np, idx_left))
        #print('this is x_phi_1[right_ind]')
        #print(tf.gather(x_phi_1, idx_right))
        #interval = [tf.gather(x_phi_np, idx_left), tf.gather(x_phi_np, idx_right)]
        #print('interval:')
        #print(interval)
        #print('interval[0]:')
        #print(interval[0])
        #print('interval[1]:')
        #print(interval[1])
        # h for x in that interval
        #int_diff = interval[1] - interval[0]  # this is positive value
        int_diff = tf.gather(x_phi_np, idx_right) - tf.gather(x_phi_np, idx_left)
        #print(int_diff)
        # Using relu. Compute the linear combination with TensorFlow ops
        interp = (tf.gather(tf.transpose(phi_0), idx_left) * (tf.nn.relu(x - tf.gather(x_phi_1, idx_left)) / (-int_diff) + 1) +
                  tf.gather(tf.transpose(phi_0), idx_right) * (tf.nn.relu(x - tf.gather(x_phi_1, idx_left)) / int_diff))
        #interp = (tf.transpose(phi_0)[idx_left] * (tf.nn.relu(x - x_phi_1[idx_left:idx_right]) / (-int_diff) + 1) +
        #          tf.transpose(phi_0)[idx_right] * (tf.nn.relu(x - x_phi_1[idx_left:idx_right]) / int_diff))

        # Sum the interpolation result and append to results list
        results.append(tf.reduce_sum(interp))

    # Convert results to a TensorFlow tensor
    results = tf.stack(results)

    return results
```

```python
# sorted takes things that need to be in order!!!
def phi1_new(x_arr, x_phi_1 = x_phi_1, phi_0 = phi_0):
    # Convert inputs to TensorFlow tensors if they aren't already
    x_arr = tf.convert_to_tensor(x_arr, dtype=tf.float64)
    x_phi_1 = tf.convert_to_tensor(x_phi_1, dtype=tf.float64)
    phi_0 = tf.convert_to_tensor(phi_0, dtype=tf.float64)
    #print(phi_0.shape)
    # Flatten the x_phi_1 tensor
    x_phi_np = tf.reshape(x_phi_1, [-1])

    results = []

    # Find the indices where x_1 would fit in the sorted array x
    for x in tf.unstack(x_arr):
        idx_left = tf.cast(tf.searchsorted(tf.constant(x_phi_np), tf.constant(x, dtype=tf.float64), side='left'), tf.int64)
        idx_right = tf.cast(tf.searchsorted(tf.constant(x_phi_np), tf.constant(x, dtype=tf.float64), side='right'), tf.int64)

        if tf.constant(x, dtype=tf.float64) <= x_phi_np[0]:  # if x is out of range, use first two indices
            idx_left = 0
            idx_right = 1
        elif tf.constant(x, dtype=tf.float64) >= x_phi_np[-1]:  # if x is out of range, use last two indices
            idx_left = len(x_phi_np) - 2
            idx_right = len(x_phi_np) - 1
        else:
            idx_left = tf.maximum(idx_left - 1, 0)
            #print(f'left index:, {idx_left}')
            idx_right = tf.minimum(idx_right, len(x_phi_np) - 1)

        # Get the 'h's by looking at interval [x[idx_left], x[idx_right]]

        int_diff = tf.gather(x_phi_np, idx_right) - tf.gather(x_phi_np, idx_left)

        # Using relu. Compute the linear combination with TensorFlow ops, tf.gather used to find index
        interp = (tf.gather(tf.transpose(phi_0), idx_left) * (tf.nn.relu(x - tf.gather(x_phi_1, idx_left)) / (-int_diff) + 1) +
                  tf.gather(tf.transpose(phi_0), idx_right) * (tf.nn.relu(x - tf.gather(x_phi_1, idx_left)) / int_diff))
        #interp = (tf.transpose(phi_0)[idx_left] * (tf.nn.relu(x - x_phi_1[idx_left:idx_right]) / (-int_diff) + 1) +
        #          tf.transpose(phi_0)[idx_right] * (tf.nn.relu(x - x_phi_1[idx_left:idx_right]) / int_diff))

        # Sum the interpolation result and append to results list
        results.append(tf.reduce_sum(interp))

    # Convert results to a TensorFlow tensor
    results = tf.stack(results)
    results = tf.reshape(results, (len(results),1))

    return results
```
